## Calendar

```python
file_calendar = '../../Data/VacationPlanner/calendar.csv.gz'
calendar = pd.read_csv(file_calendar)

calendar.head()
```

```python
calendar.describe()
```

```python
calendar.dtypes
```

```python
calendar.shape
```

```python
calendar.isnull().sum()
```

```python
calendar['date'] = pd.to_datetime(calendar['date'], yearfirst=True)
```

```python
calendar['price'] = calendar['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
calendar['available'].value_counts()
```

```python
sns.scatterplot(x="date", y="price", data=calendar, alpha=0.05)
```

```python
sns.scatterplot(x="date", y="price", data=calendar[calendar['price']<1000], alpha=0.05)
```
