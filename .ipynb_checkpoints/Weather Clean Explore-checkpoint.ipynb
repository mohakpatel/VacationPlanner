# Clean and explore weather data


Finding powder on a mountain makes skiing way more fun. While timing a ski trip on weather prediction is hard, I still want to look at the past weather history in the region to maximize my chances of finding powder. 

There are many ski resorts near Denver. I choose to analyze the weather at Vail because it is roughly in the center of all the ski resorts, so it should provide a good estimation of the weather in the region. I download the daily weather summary for past ten years at Vail from National Centers for Environmental Information. For anyone looking for weather data, I will highly recommend checking their website: https://www.ncdc.noaa.gov/cdo-web/search.

![title](../../Data/VacationPlanner/Vail.jpg)


In the weather data of past ten years, I am looking to find the trend of snowfall, snow depth and temperature in Vail over winter months, hoping to find a time frame for optimal snow conditions for skiing in Colorado. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Weather data requested

To analyze the weather trend, I request the following data for each day from the weather station in Vail. 

Requested weather data :  
SNOW - Snowfall  
TMAX - Maximum temperature  
TAVG - Average Temperature  
TMIN - Minimum temperature  
PRCP - Precipitation  
TOBS - Temperature at the time of observation  
SNWD - Snow depth  

```python
file = '../../Data/VacationPlanner/vail_daily.csv'

df = pd.read_csv(file)
df.head()
```

### Drop not extra columns

The data had some extra columns which are not useful for our purpose, so I remove them from the dataset. 

```python
df = df.drop(columns=['MDPR', 'MDSF', 'TOBS', 'WESD', 'WESF'])
```

## Missing values in dataset

Looking at the dataset, there is quite a bit of missing data. However, checking the documentation from the weather station, they do not provide any causes for the missing data. Initially, I thought that nan in SNOW and SNWD might represent no snowfall on the day the data was collected. However, nans also exist in temperature column which don't make sense. 

I thought about imputing the nan values, however, since I am only interested in looking at the average values over 10 years, I am just going to do the summary description ignoring nan values. 

```python
df.isnull().sum()
```

## Data cleaning

1. Format date into into pandas datatime format. But before doing that, I change only the year of all the past entry from ten years, to 2018-2019. This will allow me to use groupby function to compute mean and standard deviation for the coming winter from 2018 to 2019. 

2. Convert temperature from Fahrenheit to Celsius.

3. Crop the data range between the months of October till the end of May. 

```python
# Format date
def format_date(x):
    if x[5:7] in ['12', '11', '10']:
        x = '2018' + x[4:]
    else:
        x = '2019' + x[4:]
    return x

df['date'] = df['DATE'].apply(format_date)
df = df[df['date'] != '2019-02-29'] # Remove date entries from leap year
df['date'] = pd.to_datetime(df['date'], yearfirst=True)
df = df.drop(columns=['DATE'])

# Convert temp from F to C
def temp_f_to_c(temp_f):
    temp_c = (temp_f - 32) * 5 / 9
    return temp_c

df[['TAVG', 'TMAX', 'TMIN']] = df[['TAVG', 'TMAX', 'TMIN']].apply(temp_f_to_c)


# Crop data range to winterish months
df = df[(df['date'] >= '2018-10-01') & (df['date'] < '2019-06-01')]
```

## Data processing

Compute mean and standard deviation for all the weather quants on each day from past 10 years. Then, we apply 2 weeks of rolling averaging centered around each date to smooth the results. 

```python
df_stats = df.groupby('date').mean()
df_std = df.groupby('date').std()

std_columns = ['PRCP_STD', 'SNOW_STD', 'SNWD_STD', 'TAVG_STD', 'TMAX_STD', 'TMIN_STD']
std_columns_ = ['PRCP', 'SNOW', 'SNWD', 'TAVG', 'TMAX', 'TMIN']
df_stats[std_columns] = df_std[std_columns_]
df_stats = df_stats.rolling(15).mean()
```

## Weather trend

Now, we plot the weather trend in terms of temperature, snow depth and daily snowfall. The plots also show the standard deviation from the average value. 

We can see that we should be able to ski between the start of December to end of April. Based on the average trend the best time to ski looks to be between start of January to start of April. However in a good season, we should also be able to ski in month of April and begining of May.

```python
fig, ax = plt.subplots(3,1, sharex=True, figsize = (12,9))

fig.suptitle('Historical weather trends in Vail, CO', fontsize=16)

# Temperature
lower = df_stats['TAVG'] - df_stats['TAVG_STD']
upper = df_stats['TAVG'] + df_stats['TAVG_STD']
ax[0].plot(df_stats.index, 0*df_stats['TAVG'], ':k', alpha=0.2)
ax[0].plot(df_stats.index, df_stats['TAVG'], 'k', color='#fdae6b')
ax[0].fill_between(df_stats.index, lower, upper,
            alpha=0.5, edgecolor='#fdd0a2', facecolor='#fff5eb')
ax[0].spines["top"].set_visible(False)
ax[0].spines["right"].set_visible(False)
ax[0].spines["bottom"].set_visible(False)
ax[0].set_ylabel('Temperature (C)', fontdict={'size':14})



# Snow depth
lower = df_stats['SNWD'] - df_stats['SNWD_STD']
lower[lower<0] = 0
upper = df_stats['SNWD'] + df_stats['SNWD_STD']
ax[1].fill_between(df_stats.index, lower*0, df_stats['SNWD'],
            alpha=0.5, facecolor='#deebf7')
ax[1].fill_between(df_stats.index, df_stats['SNWD'], upper,
            alpha=0.5, facecolor='#f7fbff')
ax[1].plot(df_stats.index, lower, '--', color='#9ecae1')
ax[1].plot(df_stats.index, upper, '--', color='#9ecae1')
ax[1].plot(df_stats.index, df_stats['SNWD'], 'k', color='#4292c6')
ax[1].spines["top"].set_visible(False)
ax[1].spines["right"].set_visible(False)
ax[1].spines["bottom"].set_visible(False)
ax[1].set_ylabel('Snow depth (inches)', fontdict={'size':14})


# Snow fall
lower = df_stats['SNOW'] - df_stats['SNOW_STD']
lower[lower<0] = 0
upper = df_stats['SNOW'] + df_stats['SNOW_STD']

ax[2].plot(df_stats.index, df_stats['SNOW'], 'k', color='#a1d99b')
ax[2].fill_between(df_stats.index, lower, upper,
            alpha=0.5, edgecolor='#c7e9c0', facecolor='#e5f5e0')
ax[2].spines["top"].set_visible(False)
ax[2].spines["right"].set_visible(False)
ax[2].spines["bottom"].set_visible(True)
ax[2].set_ylabel('Snowfall (inches)', fontdict={'size':14});

plt.savefig('WeatherTrend.png', dpi=800)
```

## Save cleaned data for further analysis

Now, we will crop the data between 15th December and 15th April and save the average columns for further analysis.

```python
DF = df_stats[['SNOW', 'SNWD', 'TAVG', 'SNOW_STD', 'SNWD_STD', 'TAVG_STD']]
DF['date'] = DF.index
DF.reset_index(drop=True)
DF = DF[(DF['date'] >= '2018-12-15') & (DF['date'] <= '2019-04-15')]
DF.to_csv('../../Data/VacationPlanner/weather_trend.csv')
```
